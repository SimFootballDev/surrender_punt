```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import re
```

```python
requestHead = {"User-Agent": "Chrome/47.0.2526.111"}
```

```python
def getSeason(num):
    if num < 10:
        strnum = '0' + str(num)
    elif num >= 10:
        strnum = str(num)
    url = "http://sim-football.com/indexes/NSFLS"+strnum+"/GameResults.html"
    
    page = requests.get(url)
    soup = BeautifulSoup(page.text,'html.parser')
```

```python
strnum = '17'

url = "http://sim-football.com/indexes/NSFLS"+strnum+"/GameResults.html"

page = requests.get(url,  headers = {"User-Agent": "Chrome/47.0.2526.111"})
soup = BeautifulSoup(page.text,'html.parser')
```

```python
PbPList = soup.find_all('a', href = re.compile("Logs"))
PbPList = [p.get('href') for p in PbPList]
PbPList[0]
```

```python
page2 = requests.get("http://sim-football.com/indexes/NSFLS"+strnum+"/"+PbPList[0], headers = requestHead)
```

```python
soup3 = BeautifulSoup(page2.content,'lxml')
```

```python
table = soup3.find_all('table', class_='Grid')[0]
```

```python
q = -2
pbpList = []
for row in table.find_all('tr'):
    cols = row.find_all('td')
#     print(cols)
    if len(cols) != 5:
        q = q + 1
    elif len(cols) == 5:
        team = cols[0].find_all('img')[0]['src'][16]    #2 is home, 1 is away
        minutes,seconds = cols[1].text.strip().split(':')
        secondsLeft = int(minutes)*60 + int(seconds)
        tottime = 15*60*(4-q) + secondsLeft
        downDist = cols[2].text
        if downDist == '':
            down = ''
            dist = ''
        elif downDist == '---':
            down = ''
            dist = ''
        else:
            down = downDist[0]
            dist = downDist[8:]
#         print(cols[3])
        loc = cols[3].text
        if loc == '':
            side = yard = ''
        else:
            side = loc[:3]
            yard = int(loc[6:])
        play = cols[4].text
        pbpList.append((team,q,secondsLeft,tottime,down,dist,side,yard,play))
```

```python
pbpArr = np.asarray(pbpList)
df = pd.DataFrame(pbpArr)
df.columns = ['teamID','Q','time','totTime','down','distance','side','yard','play']
```

```python
teamList = df['side'].unique()
homeTeam = teamList[0]
awayTeam = teamList[-1]
```

```python
def teams(teamID,homeTeam,awayTeam):
    if teamID == '2':
        return homeTeam
    elif teamID == '1':
        return awayTeam
```

```python
df['team'] = df.apply(lambda row : teams(row['teamID'], homeTeam, awayTeam),axis=1)
```

```python
def dist2goal(team,side,yard):
    if team == side:
        dist = 100 - int(yard)
    elif side != '':
        dist = int(yard)
    else:
        dist = ''
    return dist
```

```python
def goal2go(distance,dist2goal):
    if distance == 'Goal':
        return str(dist2goal)
    else:
        return distance
```

```python
df['dist2goal'] = df.apply(lambda row : dist2goal(row['team'],row['side'],row['yard']),axis=1)
```

```python
df['distance'] = df.apply(lambda row : goal2go(row['distance'],row['dist2goal']),axis=1)
```

```python
df = df[['team','Q','time','totTime','down','distance','dist2goal','play']]
```

```python
boxList = soup.find_all('a', href = re.compile("Boxscore"))
boxList = [p.get('href') for p in boxList]
boxList[0]
```

```python
pagebox = requests.get("http://sim-football.com/indexes/NSFLS"+strnum+"/"+boxList[0], headers = requestHead)
```

```python
soupbox = BeautifulSoup(pagebox.content,'lxml')
```

```python
tablebox = soupbox.find_all('table', class_='Grid')[0]
```

```python
scoreList = [(3600,0,0)]
# scoreList = []
```

```python
q=0
for row in tablebox.find_all('tr')[1:]:
    cols = row.find_all('td')
#     print(len(cols))
    if len(cols) != 6:
        q += 1
#         scoreList.append((q,901,awayscore,homescore))
    else:
#         print(cols[-4])
        time = cols[-4].text.split(':')
        secondsLeft = int(time[0]) * 60 + int(time[1])
        totTime = 15*60*(4-q) + secondsLeft
        awayscore = int(cols[-2].text)
        homescore = int(cols[-1].text)
        scoreList.append((totTime,awayscore,homescore))
scoreList.append((0,awayscore,homescore))
```

```python
scoreList
```

```python
boxArr = np.asarray(scoreList)
boxdf = pd.DataFrame(boxArr)
boxdf.columns = ['totTime','awayScore','homeScore']
```

```python
boxdf
```

```python
def getScore(boxDF,totTime):
    i = 0
#     print('\n')
#     print(totTime)
    while int(totTime) <= int(boxDF.loc[i+1]['totTime']):
        i += 1
#         print('+1')
#     print(boxDF.loc[i]['totTime'])
#     print(boxDF.loc[i]['awayScore'],boxDF.loc[i]['homeScore'])
    return boxDF.loc[i]['awayScore'],boxDF.loc[i]['homeScore']
```

```python
df['awayScore'] = df.apply(lambda row : getScore(boxdf,row['totTime'])[0],axis=1)
```

```python
df['homeScore'] = df.apply(lambda row : getScore(boxdf,row['totTime'])[1],axis=1)
```

```python
df = df[['team','Q','time','totTime','awayScore','homeScore','down','distance','dist2goal','play']]
```
