# Surrender Index

```python
import pandas as pd
import numpy as np
```

```python
def scoreDiff(row):
    if row['recTeam'] == row['awayTeam']:
        return row['homeScore'] - row['awayScore']
    else:
        return row['awayScore'] - row['homeScore']
```

```python
def secSinceHalf(row):
    return 1800 - int(row['totTime'])
```

```python
def fieldPosScore(row):
    if row['dist2goal'] >= 60:
        return 1.0
    elif row['dist2goal'] >= 50:
        return 1.1 ** (40 - row['dist2goal'])
    else:
        return 1.1**10 * 1.2**(50 - row['dist2goal'])
```

```python
def distScore(row):
    if int(row['distance']) >= 10:
        return 0.2
    elif 7 <= row['distance'] <= 9:
        return 0.4
    elif 4 <= row['distance'] <= 6:
        return 0.6
    elif 2 <= row['distance'] <= 3:
        return 0.8
    else:
        return 1.0
```

```python
def scoreDiffMult(pointdiff):
    if pointdiff > 0:
        return 1
    elif pointdiff == 0:
        return 2
    elif pointdiff < -8:
        return 3
    else:
        return 4
```

```python
def timeMult(row):
    sec = secSinceHalf(row)
    PD = scoreDiff(row)
    
    if PD <= 0 and sec >= 0:
        return ((sec * 0.001)**3) + 1.
    else:
        return 1.
```

```python
def surrenderIndex(row):
    pd = scoreDiff(row)
#     print(fieldPosScore(row),distScore(row),scoreDiffMult(pd),timeMult(row))
    return fieldPosScore(row) * distScore(row) * scoreDiffMult(pd) * timeMult(row)
```

```python
allData = pd.read_csv('all.csv')
```

```python
puntDF = allData[allData['play'].str.contains('Punt ')]
```

```python
puntDF = puntDF.rename(columns = {'teamPoss':'recTeam'})
```

```python
def fixAZBug(row): #Only for Punts!
    if row['recTeam'] == 'AZ':
        row['dist2goal'] = 100 - row['dist2goal']
        if row['homeTeam'] == 'AZ':
            row['side'] = row['awayTeam']
        else:
            row['side'] = row['homeTeam']
    return row
```

```python
puntDF = puntDF.apply(lambda row : fixAZBug(row),axis=1)
```

```python
puntDF['surrenderIndex'] = puntDF.apply(lambda row : surrenderIndex(row),axis=1)
```

```python
puntDF = puntDF[['S','gameID','Q','time','awayTeam','awayScore','homeScore','homeTeam','recTeam','down','distance','dist2goal','surrenderIndex','play']]
```

```python
puntDF = puntDF.sort_values('surrenderIndex',ascending=False)
```

```python
from scipy.stats import percentileofscore as perc
```

```python
NFL = np.load("2009-2018_surrender_indices.npy")
```

```python
puntDF['percentiles'] = puntDF.apply(lambda row : perc(NFL,row['surrenderIndex']),axis=1)
```

```python
puntDF = puntDF[['S','gameID','Q','time','awayTeam','awayScore','homeScore','homeTeam','recTeam','down','distance','dist2goal','surrenderIndex','percentiles','play']]
```

```python
fourthDF = allData[allData['down'] == 4.0]
```

```python
def playType(play):
#     print(play)
    if 'Punt' in play:
        ptype = 'Punt'
    elif 'FG' in play:
        ptype = 'FG'
    elif 'Timeout' in play:
        ptype = 'Timeout'
    else:
        ptype = 'Try'
    return ptype
```

```python
fourthDF = fourthDF.rename(columns = {'teamPoss':'recTeam'})
```

```python
fourthDF['playType'] = fourthDF.apply(lambda row : playType(row['play']),axis=1)
```

```python
def fixAZBug2(row):
    try:
        if row['recTeam'] == 'AZ' and row['playType'] == 'Punt':
            row['dist2goal'] = 100 - row['dist2goal']
            if row['homeTeam'] == 'AZ':
                row['side'] = row['awayTeam']
            else:
                row['side'] = row['homeTeam']
        return row
    except KeyError:
        return row
```

```python
fourthDF = fourthDF.apply(lambda row : fixAZBug2(row),axis=1)
```

```python
fourthDF['playType'].value_counts(normalize=True)
```

```python
fourthArr = []
for i in range(1,100):
    i4df = fourthDF[fourthDF['dist2goal'] == i]
    fourthArr.append(i4df['playType'].value_counts(normalize=True))
```

```python
fourthArr[0]['FG']
```

```python
FGlist = []
PuntList = []
TryList = []
TOList = []
for f in fourthArr:
    try:
        FGlist.append(f['FG'])
    except KeyError:
        FGlist.append(0.)
    try:
        PuntList.append(f['Punt'])
    except KeyError:
        PuntList.append(0.)
    try:
        TryList.append(f['Try'])
    except KeyError:
        TryList.append(0.)
    try:
        TOList.append(f['Timeout'])
    except KeyError:
        TOList.append(0.)
```

```python
import matplotlib.pyplot as plt
```

```python
yardRange = range(1,100)

plt.figure(figsize=(10,5),facecolor='w')
plt.style.use('fivethirtyeight')
plt.title('Fourth Down Decisions')
plt.xlabel('Distance from Endzone')
plt.ylabel('Frequency')
plt.bar(yardRange,PuntList,width=1,label='Punts')
plt.bar(yardRange,TryList,bottom=PuntList,width=1,label='Tries')
plt.bar(yardRange,FGlist,bottom=[TryList[i]+PuntList[i] for i in range(99)],width=1,label='Field Goals')
plt.bar(yardRange,TOList,bottom=[FGlist[i] + TryList[i]+PuntList[i] for i in range(99)],width=1,label='Timeouts')
for i in [10,20,30,40,50,60,70,80,90]:
    plt.axvline(i,c='w',lw=1,alpha=0.5)
plt.xlim(0.25,99.75)
plt.ylim(0,1)
plt.legend(loc=5)
plt.tight_layout()
plt.savefig('4thdownbar.png')
plt.show()
```

```python
deadZoneDF = fourthDF[(fourthDF['dist2goal'] <= 40) & (30 <= fourthDF['dist2goal'])]
```

```python
deadZoneArr = []
for i in range(1,51):
    i4df = deadZoneDF[deadZoneDF['distance'] == i]
    deadZoneArr.append(i4df['playType'].value_counts(normalize=True))
```

```python
dFGlist = []
dPuntList = []
dTryList = []
dTOList = []
noneList = []

for f in deadZoneArr:
    i = 0
    try:
        dFGlist.append(f['FG'])
    except KeyError:
        dFGlist.append(0.)
        i += 1
    try:
        dPuntList.append(f['Punt'])
    except KeyError:
        dPuntList.append(0.)
        i+=1
    try:
        dTryList.append(f['Try'])
    except KeyError:
        dTryList.append(0.)
    try:
        dTOList.append(f['Timeout'])
    except KeyError:
        dTOList.append(0.)
        i+=1
    if i == 3:
        noneList.append(1.)
    else:
        noneList.append(0.)
```

```python
yardRange = range(1,51)

plt.figure(figsize=(10,5),facecolor='w')
plt.style.use('fivethirtyeight')
plt.title('Fourth Down Decisions - Dead Zone (Opp. 30 - 40)')
plt.xlabel('Distance from First Down')
plt.ylabel('Frequency')
plt.bar(yardRange,dPuntList,width=1,label='Punts')
plt.bar(yardRange,dTryList,bottom=dPuntList,width=1,label='Tries')
plt.bar(yardRange,dFGlist,bottom=[dTryList[i]+dPuntList[i] for i in range(50)],width=1,label='Field Goals')
plt.bar(yardRange,dTOList,bottom=[dFGlist[i] + dTryList[i]+dPuntList[i] for i in range(50)],width=1,label='Timeouts')
plt.bar(yardRange,noneList,bottom=[dTOList[i] + dFGlist[i] + dTryList[i]+dPuntList[i] for i in range(50)],width=1,label='No Data',color='lightgray')
for i in [10,20,30,40,50,60,70,80,90]:
    plt.axvline(i,c='w',lw=1,alpha=0.5)
plt.xlim(0.25,45.75)
plt.ylim(0,1)
plt.legend(loc=1)
plt.tight_layout()
plt.savefig('4thdown_deadDist.png')
plt.show()
```
